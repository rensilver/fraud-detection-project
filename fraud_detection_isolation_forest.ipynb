# Fraud Detection project

## Business problem
Financial institutions lose millions due to fraudulent transactions.<br>
Since fraud cases are rare and constantly evolving, anomaly detection methods can help identify suspicious behavior automatically.

## Dataset

Credit Card Fraud Detection:<br>
https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud

The dataset contains transactions made by credit cards in September 2013 by European cardholders.
This dataset presents transactions that occurred in two days, where we have 492 frauds out of 284,807 transactions. The dataset is highly unbalanced, the positive class (frauds) account for 0.172% of all transactions.

## Algorithm

**Isolation Forest**

Why using this algorithm?

- It's simple
- Designed for anomaly detection
- Easy to explain
- Fast to train
- Works well on tabular data

```python
class_counts = dframe["Class"].value_counts()
class_counts
```

```python
# Class imbalance in percentage
class_percentages = dframe["Class"].value_counts(normalize=True) * 100
class_percentages
```

```python
# Visualizing the Class imbalance distribution with Seaborn
plt.figure(figsize=(6,4))

ax = sns.countplot(
    data=dframe,
    x="Class"
)

total = len(dframe)

for p in ax.patches:
    percentage = f"{100 * p.get_height()/total:.2f}%"

    ax.annotate(
        percentage,
        (p.get_x()+p.get_width()/2., p.get_height()),
        ha="center",
        va="bottom",
    )

ax.set_xticklabels(["Normal", "Fraud"])

plt.title("Class Imbalance Distribution")
plt.xlabel("Transaction Type")
plt.ylabel("Count")

plt.show()
```

```python
# Distribution of transaction amounts
plt.figure(figsize=(10,5))

sns.histplot(
    data=dframe,
    x="Amount",
    bins=50
)

plt.title("Distribution of Transaction Amount")
plt.xlabel("Transaction Amount")
plt.ylabel("Frequency")

plt.show()
```

```python
# Using Log transformation for better visualization
dframe["LogAmount"] = np.log1p(dframe["Amount"])

plt.figure(figsize=(10,5))

sns.histplot(
    data=dframe,
    x="LogAmount",
    bins=50,
    kde=True
)

plt.title("Log-Transformed Transaction Amount Distribution")
plt.xlabel("Log(Transaction Amount)")
plt.ylabel("Frequency")

plt.show()
```

```python
# Comparing Fraud vs Normal Amounts
plt.figure(figsize=(10,5))

sns.boxplot(
    data=dframe,
    x="Class",
    y="Amount"
)

plt.xticks([0,1], ["Normal", "Fraud"])

plt.title("Transaction Amounts by Class")
plt.xlabel("Transaction Type")
plt.ylabel("Amount")

plt.show()
```

```python
# Correlation
corr_with_class = dframe.corr()["Class"].sort_values()

plt.figure(figsize=(8,10))

sns.heatmap(
    corr_with_class.to_frame(),
    cmap="coolwarm",
    annot=True
)

plt.title("Feature Correlation with Fraud Class")

plt.show()
```
